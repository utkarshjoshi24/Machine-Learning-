```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    
car1 = Car("audi", "a4", 2020)
car2 = Car("bmw", "m4", 2024)
print(f"Car-1:\nMake: {car1.make}\nModel: {car1.model}\nYear: {car1.year}")
print(f"\nCar-2:\nMake: {car2.make}\nModel: {car2.model}\nYear: {car2.year}")
```

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    
    def start_engine(self):
        print("Engine started")
    
car1 = Car("audi", "a4", 2020)
car2 = Car("bmw", "m4", 2024)
print(f"Car-1:\nMake: {car1.make}\nModel: {car1.model}\nYear: {car1.year}")
print(f"\nCar-2:\nMake: {car2.make}\nModel: {car2.model}\nYear: {car2.year}")
print(car1.start_engine())
```
