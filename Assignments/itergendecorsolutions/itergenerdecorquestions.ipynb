```python
n = int(input("Enter a number"))
my_list = []
for i in range(n,-1,-1):
    my_list[i] = i

print(my_list)
```

```python
def fib(n):
    if n==0:
        yield 1
    else:
        yield fib(n-1)+fib(n-2)

n = int(input("Enter a range:"))
gen = fib(n)
print(next(gen))
```
