```python
num = int(input("Enter a number"))
flag = True
if (num==0) or (num==1):
    print("Not a prime number")
else:
    for i in range(2,num):
        if(num%i==0):
            flag = False

if(flag == True): print(f"{num} is a prime number")
else: print("Not a prime number")
```

```python
def fib(num):
    if(num==0) or (num==1):
        return num
    else:
        return fib(num-1)+fib(num-2)
n = int(input("Enter a range: "))
for i in range(n):
    res= fib(i)
    print(res)
```
