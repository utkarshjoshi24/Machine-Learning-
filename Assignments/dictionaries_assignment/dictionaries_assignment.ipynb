```python
dict11 = {num: num**2 for num in range(1,11)}
new_dict11 = {}
for key,value in dict11.items():
    if (key%2==0):
        new_dict11[key] = value
print("New dictionary: ", new_dict11)
```

```python
def swapp(key, value):
    temp = key
    key = value
    value  = temp
    new_dict12[key] = value

dict12 = {num: num**2 for num in range(1,6)}
print("Dictionary before swapping: ", dict12)
new_dict12 = {}
for key,value in dict12.items():
    swapp(key,value)
print("New dictionary after swapping: ", new_dict12)
```
