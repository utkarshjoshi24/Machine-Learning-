```python
connection = sqlite3.connect("test.db")
cursor = connection.cursor()

def create_table():
    table = cursor.execute("""
    Create Table If Not Exists employees(
        id Integer Primary Key,
        name Text Not Null,
        age Integer,
        department Text
    )
    """)
    
    connection.commit()
    return table

table = create_table()
print(table)
```
