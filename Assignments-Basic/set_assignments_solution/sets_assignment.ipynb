```python
st = set([num for num in range(1,11)])
print(st)
```

```python
st.add(11)
print("11 added to set: ",st)
st.discard(1)
print("1 removed from set: ",st)
```

```python
set1 = set([num for num in range(1,6)])
set2 = set([num for num in range(1,11) if num%2==0])

print("Set 1: ", set1)
print("Set 2: ", set2)

print("Union: ", set1.union(set2))
print("Intersection: ", set1.intersection(set2))
print("Difference: ", set1.difference(set2))
print("Symmetric difference: ", set1.symmetric_difference(set2))
```
