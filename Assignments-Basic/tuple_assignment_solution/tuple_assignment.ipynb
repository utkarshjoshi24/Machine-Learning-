```python
tup = tuple([num for num in range(1,11)])
print(tup)
```

```python
print("First element: ",tup[0])
mid = int(len(tup)/2);
print("Middle element: ",tup[mid])
print("Last element: ",tup[-1])
```

```python
print("First three elements: ",tup[:3])
print("Last three elements: ",tup[-3:])
print("From index 2 to 5", tup[1:5])
```

```python
matrix = ((1,2,3),(4,5,6),(7,8,9))
print(matrix)
print("Element at 2nd row and 3rd column: ", matrix[1][2])
```

```python
tup1 = (1,2,3)
tup2 = (4,5,6)
concatenated_tup = tup1+tup2
print("New tuple: ", concatenated_tup)
```

```python
new_tup=(1,2,3,4,1,4,2,5,1,6,1)
print("Occurences of 1: ", new_tup.count(1))
print("Index of first occurence of 1: ", new_tup.index(1))
```

```python
packed_tup = (3,4,5,6,7)
a,b,c,d,e = packed_tup
print(a)
print(b)
print(c)
print(d)
print(e)
```
